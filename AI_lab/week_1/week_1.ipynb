```python
import gym
import matplotlib.pyplot as plt
import numpy as np
```

```python
def run_cartpole_steps(num_steps):
    env = gym.make('CartPole-v1')
    env.reset()
    total_reward = 0

    for step in range(num_steps):
        action = env.action_space.sample()  # Replace with your custom policy
        _, reward, done, _, _ = env.step(action)
        total_reward += reward

        if done:
            break

    env.close()
    return total_reward

def run_cartpole_episodes(num_episodes, steps_per_episode):
    env = gym.make('CartPole-v1')
    env.reset()
    episode_rewards = []

    for episode in range(num_episodes):
        total_reward = 0

        for step in range(steps_per_episode):
            action = env.action_space.sample()  # Replace with your custom policy
            _, reward, done, _, _ = env.step(action)
            total_reward += reward

            if done:
                break

        episode_rewards.append(total_reward)

    env.close()
    return episode_rewards

def plot_cumulative_rewards(rewards, title):
    cumulative_rewards = np.cumsum(rewards)
    plt.plot(cumulative_rewards)
    plt.title(title)
    plt.xlabel('Episodes')
    plt.ylabel('Cumulative Reward')
    plt.show()
```

```python
num_steps = 1000
total_reward_steps = run_cartpole_steps(num_steps)
print(f"Total reward for {num_steps} steps: {total_reward_steps}")
```

```python
num_episodes = 100
steps_per_episode = 200
episode_rewards = run_cartpole_episodes(num_episodes, steps_per_episode)
average_reward = np.mean(episode_rewards)
print(f"Average reward for {num_episodes} episodes with {steps_per_episode} steps per episode: {average_reward}")
```

```python
plot_cumulative_rewards(episode_rewards, "Cumulative Rewards for CartPole")
```
