```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv('forcast.csv')
# df = df.T
df = df.T.reset_index()
# df = df.set_index(df.iloc[0])
df = df.drop(df.index[0])
# df = df.drop(df.index[11])
df
```

```python
temp = result = pd.DataFrame({
    "product": [],
    "prediction": []
})
for i in range(0,20):
    model = Prophet(seasonality_mode='additive', yearly_seasonality=False, weekly_seasonality=False, daily_seasonality=False)
    model.add_seasonality(name='quarterly', period=4, fourier_order=6)
    router1 = pd.DataFrame()
    router1 = df[['index', i]].rename(columns={'index': 'ds', i: 'y'})
    model.fit(router1)
    future = model.make_future_dataframe(periods=1, freq='Q')  
    forecast = model.predict(future)
    print("Product:", i)
    # display(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(1))
    new_result = {"product": ("product0"+str(i)), "prediction": forecast[['yhat']].tail(1)}
    result = pd.concat([result, pd.DataFrame([new_result])], ignore_index=True)
    # fig = model.plot_components(forecast)
    # fig.show()
    forecast.to_csv('forcasts/product_'+str(i)+'.csv')
result.to_csv('results.csv')
```
